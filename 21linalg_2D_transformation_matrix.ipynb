```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# 2D input vector
v = np.array([ 3, -2 ])

# 2x2 transformation matrix
A = np.array([ [1,-1], [2,1] ])

# output vector is Av (convert v to column)
w = A@np.matrix.transpose(v)


# plot them
plt.plot([0,v[0]],[0,v[1]],label='v')
plt.plot([0,w[0]],[0,w[1]],label='Av')

plt.grid()
plt.axis((-6, 6, -6, 6))
plt.legend()
plt.title('Rotation + stretching')
plt.show()
```

```python
# 2D input vector
v = np.array([ 1, 3 ])

t=30

# 2x2 transformation matrix
A = np.array([ [math.cos(t),math.sin(-t)], [math.sin(t), math.cos(t)] ])

# output vector is Av (convert v to column)
w = A@np.matrix.transpose(v)


# plot them
plt.plot([0,v[0]],[0,v[1]],label='v')
plt.plot([0,w[0]],[0,w[1]],label='Av')

plt.grid()
plt.axis((-6, 6, -6, 6))
plt.legend()
plt.title('Rotation only using transformation matrix')
plt.show()
```

As we can see the norm/magnitude is not changed , only the rotation

```python
#Another example


# 2D input vector
v = np.array([ 1, 2 ])

t=30

# 2x2 transformation matrix
A = np.array([ [2,1], [2, 3] ])

# output vector is Av (convert v to column)
w = A@np.matrix.transpose(v)


# plot them
plt.plot([0,v[0]],[0,v[1]],label='v')
plt.plot([0,w[0]],[0,w[1]],label='Av')

plt.grid()
plt.axis((-6, 6, -6, 6))
plt.legend()
plt.title('Rotation + stretching')
plt.show()
```

Based on the above example, scalar when multiplied to a vector will increase or decrease magnitude

In the above example,

Av = lambda(v)

when the vector was multiplied , size of vector increased, which means the matrix increased the size of resultant matrix same as scalar. 

Above equation is called Eigen value equation. When it is true, then vector is called eigen vector and scalar is eigen scalar.
