Matrices - square, rectangular, symmetric, skew-symmetric, identity, zero, diagonal, triangular, augmented, complex

square matrix = is a matrix with same number of rows and columns. M*M dimension

Rectangular - is a matrix where rows and columns are not equal. M*N dimensions.

Symmetric - matrix mirrored across diagonal. Only a square matrix can be symmetric

skew-symmetric -  is a matrix with mirrored numbers across diagonal with opp signs and with diagonals being zeros.

identity matrix - all ones on diagonal and rest zeros off-diagonal.

zero - is a matrix with all elements being zero.

diagonal matrix is where all elements are zero except for diagonal elements which can be zero ro non-zero.
If diagonal elements are same then it can be written as lambda*I

Triangular matrix - is where all elememts above or below can be zero. if upper elements are zero then it is called lower triangular and if lower elements are zero, then it is upper triangular. 

Augmented or concatenated matrix - 2 matrices with same number of rows can be concatenated. result of matrix is a augmented matrix. 


```python
import numpy as np
from numpy.random import randint
```

```python
#S is a square matrix

S = randint(1,100,25).reshape(5,5)
S
```

```python
#R is a rectangular matrix

R = randint(1,100,20).reshape(5,4)
R
```

```python
#create Identity matrix

I = np.eye(5)
I
```

```python
Z = np.zeros((3,5))
Z
```

```python
#Create a diagnoal matrix with a list

D = np.diag([ 1, 2, 3, 5, 2 ])
D
```

```python
#convert matrix S to triangular matrix

print(S)
print()

U = np.triu(S)
print(U)
print()

L = np.tril(S)
print(L)
```

```python
#concatenate matrices S and R

print(S)
print()
print(R)
print()

C = np.concatenate((S,R),axis=1)
C
```
