# Vector-vector multiplication: the dot product
---


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
## many ways to compute the dot product

v1 = np.array([ 1, 2, 3, 4, 5, 6 ])
v2 = np.array([ 0, -4,  -3, 6, 5 ,3])

# method 1
dp = sum( np.multiply(v1,v2) )

print(dp)
```

```python
# method 2
dp = np.dot( v1,v2 )

print(dp)
```

```python
# method 3
dp = np.matmul( v1,v2 )

print(dp)
```

```python
# method 4
dp = 0  # initialize

# loop over elements
for i in range(0,len(v1)):
    
    # multiply corresponding element and sum
    dp = dp + v1[i]*v2[i]
    
print(dp)
```

Now plotting the dot product in a similar way for 3D vectors as it is not possible to visualize any higher dimension than 3d on a plot

```python
# two vectors in R2
v1 = np.array([ -4, -2 ,3])
v2 = np.array([ 2,  4 ,-1])

v3 = v1 * v2
print('v3: ',v3)


# plot in 3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]])

# make the plot look nicer

plt.title(r'3D vector from (0,0,0) to v3=v1*v2')
ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')

plt.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b',label='v1')
plt.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'g',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],[0, v3[2]],'r--',label='v3=v1*v2')

plt.legend()

plt.grid()
plt.show()
```

Resultant dot product vector is in red color v3[-8 -8 -3]
