```python
# two vectors in R2
v1 = np.array([ -4, -2 ,3])
s = -2.5

v3 = v1 * s
print('v3: ',v3)


# plot in 3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]])

# make the plot look nicer

plt.title(r'3D vector from (0,0,0) to v3=v1+v2')

ax.plot([0, 0],[0, 0],[-10, 10],'k--')
ax.plot([0, 0],[-10, 10],[0, 0],'k--')
ax.plot([-10, 10],[0, 0],[0, 0],'k--')

plt.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'g',label='v1')
plt.plot([0, v3[0]],[0, v3[1]],[0, v3[2]],'r',label='v3=v1*s')

plt.legend()

plt.grid()
plt.show()
```
