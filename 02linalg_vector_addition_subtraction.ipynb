```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
v1 = np.array([ 3, -1 ])
v2 = np.array([ 2,  4 ])

v3 = v1 + v2
print(v3)
```

```python

# two vectors in R2
v1 = np.array([ -4, -2 ])
v2 = np.array([ 2,  4 ])

v3 = v1 + v2
print('v3: ',v3)

# plot them
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0, v2[0]],[0, v2[1]],'r',label='v2')
#plt.plot([0, v2[0]]+v1[0],[0, v2[1]]+v1[1],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],'k',label='v3=v1+v2')

plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```

```python

# two vectors in R2
v1 = np.array([ -4, -2 ])
v2 = np.array([ 2,  4 ])

v3 = v1 + v2
print('v3: ',v3)

# plot them
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0, v2[0]]+v1[0],[0, v2[1]]+v1[1],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],'k',label='v3=v1+v2')

plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```

```python

# two vectors in R2
v1 = np.array([ -4, -2 ,3])
v2 = np.array([ 2,  4 ,-1])

v3 = v1 + v2
print('v3: ',v3)


# plot in 3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]])

# make the plot look nicer

plt.title(r'3D vector from (0,0,0) to v3=v1+v2')
ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')

plt.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b',label='v1')
plt.plot([0, v2[0]]+v1[0],[0, v2[1]]+v1[1],[0, v2[2]]+v1[2],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],[0, v3[2]],'k',label='v3=v1+v2')

plt.legend()

plt.grid()
plt.show()
```
