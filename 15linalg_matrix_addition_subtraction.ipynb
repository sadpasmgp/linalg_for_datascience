If A,B and C are matrices,

A+B = B+A (matrix addition is commutative)

A+(B+C = (A+B)+C) (matrix addition is distributive)

However Matrix multiplication is not commutative or distributive.

Shifting inflates a matrix from being a degenerate plane towards a sphere.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```python
#Create 2 matrices A and B

A = randint(1,200,16).reshape(4,4)
B = randint(1,100,16).reshape(4,4)
```

```python
A+B
```

```python
B+A
```

```python
A+B == B+A 
```

```python
#let C be A-B
C = A-B
```

```python
A+(B+C)
```

```python
(A+B)+C
```

```python
A+(B+C) == (A+B)+C
```

```python
# "shifting" a matrix

l = .3 # lambda

N = 5  # size of square matrix

D = np.random.randn(N,N) # can only shift a square matrix
print(D)
print()
Ds = D + l*np.eye(N)
print(Ds)
```

As we can see other than diagonal elements, all other elements are same
