```python
##creating vectors and plotting them


# 2-dimensional vector
v2 = [ 3, -2 ]

# 3-dimensional vector
v3 = [ 4, -3, -4 ]

# row to column (or vice-versa):
v3t = np.transpose(v3)


# plot 2D vector from (0,0) to [3,-2]  

plt.plot([0,v2[0]],[0,v2[1]])
#plt.axis('equal')
plt.plot([-4, 4],[0, 0],'k--')
plt.plot([0, 0],[-4, 4],'k--')
plt.grid()
plt.axis((-4, 4, -4, 4))
plt.title('2D vector from (0,0 to (3,-2)')
plt.show()


# plot the 3D vector from 0,0,0 to 4,-3,-4

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]])


# make the plot look nicer
#plt.axis('equal')
plt.title(r'3D vector from (0,0,0) to [ 4, -3, -4 ]')
ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')
plt.show()
```
