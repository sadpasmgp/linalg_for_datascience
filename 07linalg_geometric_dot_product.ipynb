#The dot product from a geometric perspective
---


dot product between 2 vectors, alpha = |a||b|cos(theta)

theta = acos(alpha/|a||b|)

where theta = angle between vectors a and b
alpha = dot product

if theta<90, then 2 vectors meet at acute angle=> alpha > 0
if theta>90, then 2 vectors have obtuse angle and dot product, alpha < 0
if theta=90, then dot product = 0, (orthogonal)
if theta = 0, then 2 vectors are co-incide, which mean dotproduct = a*b(product of vector length)
if theta = 180, then 2 vectors are in opp direction, means dotproduct will be -(a*b)

alpha = (a.T)*b = sigma(i=1 to n)(ab)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

# compute the angle (radians) between two vectors
ang = np.arccos( np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)) )


# draw them
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()
```

```python
## equivalence of algebraic and geometric dot product formulas

# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])


# algebraic
dp_a = np.dot( v1,v2 )

# geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

# print dot product to command
print(dp_a)
print(dp_g)
```
