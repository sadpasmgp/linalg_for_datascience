A subspace is a region of space that you can reach by linear combination of vectors, A span is set of vectors is all possible linear weighted combinations of that set. It also stretches to infinity.

v = [1 2 0], w = [3 2 1]
s = {[1 1 0],[1 7 0]}

The vectors v and w are spaid to be in span of s if they can be expressed as a linear combination of s.

Ex: vector v can be expressed as [1 2 0] = 5/6[1 1 0]+ 1/6[1 7 0]
hence v is in span of s while w is not.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# set S
S1 = np.array([1, 1, 0])
S2 = np.array([1, 7, 0])

# vectors v and w
v = np.array([1, 2, 0])
w = np.array([3, 2, 1])

# draw vectors
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, S1[0]],[0, S1[1]],[0, S1[2]],'r')
ax.plot([0, S2[0]],[0, S2[1]],[0, S2[2]],'r')

ax.plot([0, v[0]],[0, v[1]],[0, v[2]],'k')
ax.plot([0, w[0]],[0, w[1]],[0, w[2]],'b')

# now draw plane
xx, yy = np.meshgrid(range(30), range(30))
cp = np.cross(S1,S2)
z1 = (-cp[0]*xx - cp[1]*yy)*1./cp[2]
ax.plot_surface(xx,yy,z1)

plt.show()
```
