# Vector Hadamard multiplication
---


Also called element wise multiplication.
it is valid only for 2 vectors with same number of elements

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
w1 = np.array([1,3,5])
w2 = np.array([3,4,2])
```

```python
#hadamard multiplication
w3 = np.multiply(w1,w2)
print(w3)
```

```python
#dot product

w3 = np.dot(w1,w2)
w3
```

As we can see, hadamard multiplication results in a resultant vector, while dot product resultas in a scalar

# Vector outer product
---

Outer product = v.wT

Dot product = vT.w

outer product can be computed for column vectors with different shapes

```python
#notice the difference between all 3 different vector multiplications

v1 = np.array([  1, 2, 3 ])
v2 = np.array([ -1, 0, 1 ])

# outer product
print(np.outer(v1,v2))
print(np.dot(v1,v2))
print(np.multiply(v1,v2))
```

```python
# terrible programming, but helps conceptually:

v1 = np.array([  1, 2, 3 ])
v2 = np.array([ -1, 0, 1 ])


op = np.zeros((len(v1),len(v1)))
for i in range(0,len(v1)):
    for j in range(0,len(v2)):
        op[i,j] = v1[i] * v2[j]

print(op)
```

```python
#dot product results in scalar
#outer product results in a matrix
#hadamard multiplications results in a resultant vector
```

# Vector cross product
---

It is defined only for 2 3D vectors, resultant is also a 3D vector


[1 2 3] X [a b c] = [2c-3b  3a-1c  1b-2a]

```python
# create vectors
v1  = [ -3,  2, 5 ]
v2  = [  4, -3, 0 ]

# Python's cross-product function
v3a = np.cross( v1,v2 )
print(v3a)

# "manual" method
v3b = [ [v1[1]*v2[1] - v1[2]*v2[1]],
        [v1[2]*v2[0] - v1[0]*v2[2]],
        [v1[0]*v2[1] - v1[1]*v2[0]] ]

fig = plt.figure()
ax = fig.gca(projection='3d')

# draw plane defined by span of v1 and v2

xx, yy = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]
ax.plot_surface(xx,yy,z1)

## plot the two vectors
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'r')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')
ax.plot([0, v3a[0]],[0, v3a[1]],[0, v3a[2]],'r')


ax.view_init(azim=150,elev=45)
plt.show()
```
